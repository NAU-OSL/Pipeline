```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics import f1_score, average_precision_score, precision_recall_fscore_support

```

```python


def metric():
    columns = ['Accuracy','F-Score','AUC-PR']

    #predictions = pd.read_csv('./data/predict_file1x4.csv',header=0,dtype=int,index_col=0)
    predictions = pd.read_csv('./data/predict_file_2x2NBSStop_Header.csv',header=0,dtype=int,index_col=0)
    y_pred = predictions.values
    probabilities = pd.read_csv('./data/probability_file_2x2NBSStop_Header.csv',header=0,dtype=float,index_col=0)
    y_proba = probabilities.values

    true = pd.read_csv('./data/PRClassesNoBlankSpaces.csv',sep=';',index_col=0,header=None).loc[predictions.index]
    values = [x[0].split('-') for x in true.values]
    mlb = MultiLabelBinarizer()
    y_true = mlb.fit_transform(values)
    
    #Accuracy
    acc_ml = accuracy_ml(y_pred,y_true)
    #F1-Score
    #fmeasure_score = f1_score(y_pred,y_true, average='micro')
    fmeasure_score = f1_score(y_true,y_pred, average='macro')
    #AUC-PR
    pr_score = average_precision_score(y_true,y_proba,average='macro')
    
    prec, rec, fbeta, sup = precision_recall_fscore_support(y_true,y_pred, average='macro')

    arr = [acc_ml,fmeasure_score,pr_score]
    
    print(arr)
    
    columns2 = ['Precision','Recall','fbeta_measure','support']
    
    arr2 = [prec, rec, fbeta, sup]
    
    print(arr2)

    df = pd.DataFrame([arr],columns=columns)

    print(df)

def accuracy_ml(y_pred,y_true):
    y = y_true == 1
    y_t = [[index for index, value in enumerate(data) if value == 1] for data in y_true]
    y_p = [[index for index, value in enumerate(data) if value == 1] for data in y_pred]
    acc = 0
    for i in range(len(y_t)):
        s,t = set(y_t[i]),set(y_p[i])
        union = s.union(t)
        intersection = s.intersection(t)
        acc += (len(intersection)/len(union))
    return acc/len(y_t)

```
